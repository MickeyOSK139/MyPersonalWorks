# Linear Least-Squares Fit Part 1

Say you’ve collected the following data by making measurements at 7 different x locations. You
did each measurement many times and found the mean and standard error at each point, as
described above. The data (means) with standard errors turned out to be:
 - Make a plot of the data as points with error bars using the errorbar plotting function

```python
xdata = np.array ([ 0 , 10 , 20 , 30 , 40 , 50 , 60])
ydata = np.array ([ 43.5 , 79.4 , 156.0 , 326.1 , 196.5 , 368.1 , 412.7])
yerr = np.array ([ 30.7 , 13.6 , 79.9 , 65.2 , 84.2 , 16.1 , 27.2])
```

```python
plt.errorbar(xdata, ydata, yerr)
plt.title('A plot with error bars')
plt.legend()
```

```python
from scipy.optimize import curve_fit
# Define the function to fit.
# The first parameter is the independent variable .
# The remaining parameters are the fit parameters .
def line (x ,a , b ) :
    return a *x + b
# Do the fit. Read the help on curve_fit !
( popt , pcov ) = curve_fit ( line , xdata , ydata , sigma = yerr , absolute_sigma = True )
# popt now holds the optimized parameters a in popt [0] and b in popt [1]
# pcov is the covariance matrix , which gives errors and correlations .
# To extract just the errors on the fit parameters as sigmas :
perr = np.sqrt ( np.diag ( pcov ) )
# now perr holds the + - 1 sigma error for a and b.
```

```python
# First plot the data with error bars
plt.errorbar ( xdata , ydata , yerr , fmt ="o", label =" data ")
# There are many ways to plot the line given a and b, but here ’s one:
yfit = line ( xdata , * popt ) # * passes a list as remaining function parameters
plt.plot ( xdata , yfit , label =" fit")
plt.legend ()
plt.title ("Data along the linear fit curve")
plt.savefig ("hw1_meaningful_file_name.pdf") # PDFs can be used as LaTeX figures
```

Calculate the residuals and plot those with error bars: residuals = ydata - yfit. It
sometimes helps to draw a horizontal line at y = 0: hlines(0, min(xdata), max(xdata)).
If everything is working, the line should be within 1 sigma of the data point 68% of time.
In other words, the line should cross through about 2/3 of the error bars.

```python
residuals = ydata - yfit
plt.errorbar ( xdata , residuals , yerr , fmt ="o", label =" data ")
plt.hlines(0, 0, np.max(xdata))
plt.title ("Residual Plot")
```

Calculate and plot the normalized residuals with normalized error bars (normalized error
bars are all ±1). normalized residuals = residuals/yerr. Again, draw a horizontal
line through zero. Again, if everything is working, the line should be within 1 sigma 68% of
time. These plots help you see if you model is qualitatively the right one.

```python
plt.errorbar ( xdata , residuals/ yerr , yerr/ yerr , fmt ="o", label =" data ")
plt.hlines(0, 0, np.max(xdata))
plt.title ("Normalized Residual Plot")
```

# Linear Least-Squares Fit Part 2

Calculate $\chi^2$
, the sum of the squares of all the normalized residuals:

```python
chi_squared = np.sum((residuals/ yerr)**2)
print("the chi squared is", chi_squared)
```

 Calculate the reduced $\chi^2$
(also known as $\chi^2$ per degree of freedom).

```python
M = len( xdata ) # number of data points
df = M - len( popt ) # M -2 degrees of freedom : M data points minus 2 parameters
reduced_chi_squared = chi_squared / df
print("the reduced chi squared is", reduced_chi_squared)
```

Look up the probability that you’d randomly have found a $\chi^2$ of this size or bigger.

```python
from scipy.stats import chi2
PTE = 1 - chi2.cdf ( chi_squared , df = df ) # Probability to Exceed
print ('the prob is', PTE*100, '%')
```

# Linear Least-Squares Fit Part 3
Do the entire fitting process of Problems 4-5 at least 10000 times (maybe 1000 while you’re
debugging), as follows. Start by playing god and picking some particular physical constants
for the true slope a (say 7) and true intercept b (say 50). Keep the same xdata and yerr.
Initialize any variables that you need to record things as empty lists. Now make a loop, and
each time through your loop:

```python
def generate_y_data(xdata, yerr, a, b):
    size = len(xdata)
    true_y = np.zeros (size)
    for k in range (size):
        true_y[k] = a*xdata[k] + b
    ydata = np.zeros (size)
    for k in range (size):
        ydata[k] = np.random.normal(loc=true_y[k], scale=yerr[k])
    return ydata 
```

```python
xdata = np.array ([ 0 , 10 , 20 , 30 , 40 , 50 , 60])
yerr = np.array ([ 30.7 , 13.6 , 79.9 , 65.2 , 84.2 , 16.1 , 27.2])
size = len(xdata)
a = 7
b = 50
alist, blist, perrlist, chisqlist, redchisqlist, PTElist = [], [], [], [], [], []
for i in range (10000):
    ydata = generate_y_data(xdata, yerr, a, b)
    ( popt , pcov ) = curve_fit ( line , xdata , ydata , sigma = yerr , absolute_sigma = True )
    perr = np.sqrt(np.diag(pcov))  
    yfit = line ( xdata , * popt )
    residuals = ydata - yfit
    chi_squared = np.sum((residuals/ yerr)**2)
    reduced_chi_squared = chi_squared/(size-2)
    PTE = 1 - chi2.cdf(chi_squared, df = (size-2) )


    alist.append(popt[0])
    blist.append(popt[1])
    perrlist.append(perr)
    chisqlist.append(chi_squared)
    redchisqlist.append(reduced_chi_squared)
    PTElist.append(PTE)
```
